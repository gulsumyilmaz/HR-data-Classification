```python
dolu = df[df['Attrition'].notnull()] # we got the empty values
```

```python
dolu.shape
```

```python
x=dolu.drop("Attrition",axis=1)
y=dolu["Attrition"]
```

```python
x
```

```python
df["Department"].unique()
```

```python
x=pd.get_dummies(x,drop_first=True) #data model ready
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
```

```python
g= GaussianNB()
b= BernoulliNB()
```

```python
g.fit(x,y)
```

```python
b.fit(x,y)
```

```python
#models trained
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score,confusion_matrix,classification_report
```

```python
tahmin1=g.predict(x)
tahmin2=b.predict(x)
```

```python
accuracy_score(tahmin1,y)
```

```python
accuracy_score(tahmin2,y)
```

```python
precision_score(tahmin2,y)
```

```python
confusion_matrix(tahmin2,y) #True Positive, False Negative, False Positive, True Negative
```

```python
#962 people stay at work is the right guess, 56 people quit is the right guess
#other 2 value the wrong guess
```

```python
confusion_matrix(tahmin1,y)
```
